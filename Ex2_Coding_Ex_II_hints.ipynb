### In the following, you will find an example plotting code using dummy data. To afterwards plot the real wind data, use a for loop to loop over the months and create one plot for each month.

```python
# An example of a plotting code to start with 

# Create the coordinates
x = np.linspace(0, 10, 15)
y = np.linspace(0, 10, 15)
X, Y = np.meshgrid(x, y)

# Zonal and meridional wind components (example values, replace with your data)
u = np.cos(X)  # Dummy data for the vertical wind component
v = 0.1*X  # Dummy data for the meridional wind component

# Plotting
plt.figure(figsize=(8, 6))

# Add wind vectors (quiver plot)
plt.quiver(X, Y, u, v, scale=30, color='black')  # Adjust scale to make arrows finer

# Add labels and title
plt.xlabel('X (Longitude)')
plt.ylabel('Y (Latitude)')
plt.title('Wind Direction')

# Show plot
plt.show()
```

```python
# Only plot locations where the wind vecotor is westerly (thus, u>0)

# Plotting
plt.figure(figsize=(8, 6))

westerly_mask = np.where(u>0,1,np.nan) # Matrix with ones and NaNs, masking where winds are not westerly. Multiply this matrix point-wise with the wind matrices in order to get a masked matrix

print('Masking matrix:')
print(westerly_mask)

print('Original zonal wind component: ')
print(u)

print('Masked zonal wind component (Only containing locations with westerly wind, every other location set to NaN): ')
print(u*westerly_mask)


# Plotting
plt.figure(figsize=(8, 6))

# Add wind vectors (quiver plot). Hint: Plot only every 4th point in order to avoid having a very populated plot.
plt.quiver(X, Y, westerly_mask*u, westerly_mask*v, scale=30, color='black')  # Adjust scale for vector size


# Add labels and title
plt.xlabel('X (Longitude)')
plt.ylabel('Y (Latitude)')
plt.title('Masked Wind  Direction')

# Show plot
plt.show()
```
