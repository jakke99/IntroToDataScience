```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#this is a nested list

july = np.array([
    [1, 5, 35],     # Site 1
    [14, 2, 0],     # Site 2
    [0, 31, 67],    # Site 3
    [96, 110, 78],  # Site 4
    [0, 0, 0]       # Site 5
])

august = np.array([
    [15, 23, 10],    # Site 1
    [54, 96, 240],   # Site 2
    [0, 3, 9],       # Site 3
    [12, 31, 27],    # Site 4
    [8, 14, 6]       # Site 5
])

september = np.array([
    [48, 78, 170],   # Site 1
    [2, 0, 0],       # Site 2
    [0, 11, 14],     # Site 3
    [25, 13, 12],    # Site 4
    [131, 96, 43]    # Site 5
])
```

```python
sum_all = august + september + july
print("Total number of fish:\n"  + str(sum_all))
```

```python
#Efficiancy of the net is 0.5 for SP2 and SP3 0.25

efficiency = np.array([
    [1],
    [2],
    [4]
])

corrected_matrix = np.matmul(july, efficiency)

print("Corrected per site in July :\n" + str(corrected_matrix))

total_corrected = np.sum(corrected_matrix)

print("Total number of fish:" + str(total_corrected))
```

```python
# List of month names and corresponding data arrays
months = ['july', 'august', 'september']
data_arrays = [july, august, september]

# Efficiency vector (as a 1D array)
efficiency  = np.array([
    [1],
    [2],
    [4]
])

for name, data in zip(months, data_arrays):
    corrected = np.matmul(data, efficiency)
    print(f"Corrected per site in {name.capitalize()}:\n{corrected}")
    print(f"Total corrected fish in {name.capitalize()}: {np.sum(corrected)}\n")
```

```python
#Rewrite this to have it for all months 

X, Y = np.meshgrid(lon, lat)
fig, axes = plt.subplots(3, 4, figsize=(20, 12))
axes = axes.flatten()

for i in range(12):
    ax = axes[i]

    # Select months
    u_month = u[i, :, :]
    v_month = v[i, :, :]

    ax.quiver(X, Y, u_month, v_month, scale=1000, color='black')

    ax.set_title(months[i])
    plt.xlabel('Longitude (°E)')
    plt.ylabel('Latitude (°N)')

fig.suptitle('Monthly Mean Wind Vectors (2023)', fontsize=16)
plt.tight_layout()
plt.show()


```
