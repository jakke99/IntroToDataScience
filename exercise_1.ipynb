```python
#Variables
np.pi
np.cos(np.pi)
A=1
A=2
T=12
t0=6
C=3

#t = np.arange(0,36.1,1)
t = np.arange(0,36.1,6.13)
#Change Timesteps: 
#(0.25, 1,3.47,6.13))
print(t)
```

```python
def func(t):
    return A*np.cos(2*np.pi / T * (t-t0)) + C

g_t = func(t)


print(g_t)
```

```python
fig = plt.figure()
plt.plot(t,func(t))
plt.grid()

plt.xlabel('t',fontsize=20)
plt.ylabel('g(t)',fontsize=20)
plt.show()
```

```python
#Compute mean for one timestep

mean_g_t = np.mean(g_t)
print("Mean g(t):", mean_g_t)
```

```python
# Make it beautiful with list: 


# Different time step sizes
dt_values = [0.25, 1, 3.47, 6.13]

for dt in dt_values:
    t = np.arange(0, 36 + dt, dt)

    g_t = func(t)

    mean_val = np.mean(g_t)
    print(str(dt) + "=>" + str(mean_val))
```

```python
# exercise 3

def h(x, mu, sigma):

    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
```

```python
# Parameters
mu = 0       # mean
sigma = 1    #Standart Dev


x = np.linspace(-5, 5, 500)


h_x = h(x, mu, sigma)
```

```python
plt.plot(x, h_x,)
plt.title('')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#b

def sums_and_products(n, m):
   
    
    arr = np.arange(n, m + 1)
    
    # (a) sum of i
    sum_i = np.sum(arr)
    
    # (b) sum of (3j - 2)
    sum_j = np.sum(3*arr - 2)
    
    # (c) product of i
    prod_i = np.prod(arr)
    
    # (d) product of 2k
    prod_k = np.prod(2*arr)
    
    # Print results
    print(f"(a) Î£ i       = {sum_i}")
    print(f"(b) Î£ (3j-2)   = {sum_j}")
    print(f"(c) Î  i       = {prod_i}")
    print(f"(d) Î  2k      = {prod_k}")
    
    return sum_i, sum_j, prod_i, prod_k
```

```python
sums_and_products(2, 5)
```
